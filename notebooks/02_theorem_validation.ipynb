# Notebook 02 — Theorem Validation

Interactively verifies all theorems from Section IV of the paper.


```python
from abt import build_tree_from_list
from abt.math import (
    verify_theorem1, verify_theorem2, verify_corollary3,
    verify_theorem4, verify_theorem5,
    verify_property1, verify_property2,
    print_verification_report,
)

TEST_CASES = [
    ([10, 6, 14, 4, 8, 12, 16], 'h=2 canonical'),
    ([20, 10, 30],               'h=1 minimal'),
    ([100, 50, 150, 25, 75, 125, 175, 10, 40, 60, 90, 110, 140, 160, 190], 'h=3'),
    ([1.0, 0.5, 1.5, 0.25, 0.75, 1.25, 1.75], 'float values'),
]

for values, label in TEST_CASES:
    T = build_tree_from_list(values)
    print_verification_report(T, label=label)
```
