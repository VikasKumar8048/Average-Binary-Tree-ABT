# Notebook 01 — ABT Basics

Reference: Vikas Kumar, *Average Binary Tree (ABT)*, IJIRT Vol 13 Issue 1, June 2026.

## Contents
1. Building a perfect binary tree
2. Applying the ABT operator
3. Verifying the canonical paper examples (Figures 1 and 2)
4. Iterative contraction sequence


```python
from abt import ABT, build_tree_from_list, build_abt
from abt.viz import ascii_tree, side_by_side_ascii
```

## 1. Figure 1: Height-2 example from the paper

```python
# Paper Figure 1: T (h=2, n=7), values [10, 6, 14, 4, 8, 12, 16]
T = build_tree_from_list([10, 6, 14, 4, 8, 12, 16])
T_prime = build_abt(T)
print(side_by_side_ascii(T, T_prime))
print()
print('ABT root =', T_prime.val, '= (6+14)/2')
print('ABT left =', T_prime.left.val, '= (4+8)/2')
print('ABT right=', T_prime.right.val, '= (12+16)/2')
```

## 2. Figure 2: Height-3 example from the paper

```python
# Paper Figure 2: T (h=3, n=15)
vals_h3 = [100, 50, 150, 25, 75, 125, 175, 10, 40, 60, 90, 110, 140, 160, 190]
T3 = build_tree_from_list(vals_h3)
T3p = build_abt(T3)
print(ascii_tree(T3,  'T  (h=3, n=15)'))
print(ascii_tree(T3p, "T' (h=2, n=7)"))
```

## 3. Iterative Contraction Sequence (Corollary 3)

```python
tree = ABT.from_list(vals_h3)
seq = tree.iterative_sequence()
for i, t in enumerate(seq):
    print(f'f^{i}(T): h={t.height}, n={t.size:2d}, values={[round(v,1) for v in t.values()]}')
```

## 4. Try your own tree

```python
# A perfect binary tree of height 1 (3 nodes)
my_tree = ABT.from_list([100.0, 80.0, 120.0])
my_result = my_tree.transform()
my_tree.print('My tree')
my_result.print('My ABT')
print('Theorems pass:', my_result.verify(verbose=True))
```
